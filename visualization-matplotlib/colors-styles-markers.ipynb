#### Combining Colors, Line Styles, and Markers

* Plot the function y=(-0.3x)cos(2x)
* Line style: green
* Line style: dash-dot(-.)
* Marker: squares (s)

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10 with 50 points
x = np.linspace(0, 10, 50)

# Computer y values for the function y = e^(-0.3x) * cos(2x)
y = (-0.3*x) * np.cos(2*x)

# Plot with green color, dash-dot linestyle, and square markers
plt.plot(x, y, color='green', linestyle='-.', marker='x', label='y = e^(-0.3x) cos(2x)')

# Adding labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Damped Cosine Wave with Custom Style')
plt.legend()

plt.show()
```
