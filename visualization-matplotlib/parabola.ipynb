```python
import matplotlib.pyplot as plt
import numpy as np

# linspace generates values between start point, and endpoint and the number of points we want to have.
x = np.linspace(-10, 10, 100)
y = x**2

# plt.plot() creates a 2d plot, x is the horizontal values, y is the vertical values
plt.plot(x, y, color='RED') 

# Adding labels and title
plt.xlabel('X-axis') # Label for x-axis
plt.ylabel('Y-axis') # Label for y-axis
plt.title("Parabola: y = x^2") # Title of the plot

# Display the plot
plt.show()
```
