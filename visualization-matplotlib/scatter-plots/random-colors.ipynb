```python
import matplotlib.pyplot as plt
import numpy as np

# Coloring Markers Based on Values
# Instead of a single color, we can assign a colormap (cmap) based on another variable
# Generate random values for coloring

colors = np.random.rand(50) # Each point gets a unique color value / generates 50 random numbers between 0 and 1
x = np.random.rand(50) # Randon x-coordinates
y = np.random.rand(50) # Random y-coordinates

# Create scatter plot with colormap
plt.scatter(x, y, s=150, c=colors, cmap='viridis', alpha=0.8) # These values act as color indices for the colormap (cmap).

# When used in plt.scatter(), Matplotlib maps these values onto the 'virdis' colormap, assigning a unique color to each point
# 'viridis', 'plasma', 'inferno', 'magma', 'cividis' could be the values of cmap

plt.colorbar(label='Color Scale') # Add a color bar to show scale
plt.show()
```
