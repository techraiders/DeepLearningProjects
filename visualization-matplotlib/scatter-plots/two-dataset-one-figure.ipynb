#### Plotting two datasets in one scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,5))

# Generate first dataset
x1 = np.random.rand(50)
y1 = np.random.rand(50)

# First scatter plot
plt.scatter(x1, y1, s=10, color='red', alpha=0.7, label='Dataset 1')

# Generate second dataset
x2 = np.random.rand(50)
y2 = np.random.rand(50)

# First scatter plot
plt.scatter(x2, y2, s=10, color='blue', alpha=0.7, label='Dataset 2')

plt.legend()
plt.show()
```
