#### Customizing Markers
* Plot the function
* y = sin(x), with blue circular markers at each data point

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10 (20 points)
x = np.linspace(0, 10, 20)

# Computer y values for the function y = sin(x)
y = np.sin(x)

plt.plot(x, y, marker='o', linestyle='-', color='blue') #v, o, x, *, p, s, <. >
# Solid line: '-'
# Dashed line: '--'
# Dotten line: ':'
# Dash-dot line: '-.'

# Adding labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Sine Wave with Markers')

# Display the plot
plt.show()
```
