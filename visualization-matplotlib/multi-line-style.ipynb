#### Using different line styles
* Plot two lines on the same graph
* y = 2x + 1 using a dashed line.
* y = x + 5 using a dotten line.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from -5 to 5
x = np.linspace(-5, 5, 50)

# Define two functions
y1 = 2*x + 1 # First function
y2 = -x + 5 # Second function

# Plot the first function with a dashed line
plt.plot(x, y1, linestyle="--", label="y = 2x + 1")
plt.plot(x, y2, linestyle=":", label="y = -x + 5")

# Adding labels, title, and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Styles Example")
plt.legend() # Show legend to differentiate the lines

# Display the plot
plt.show()
```
