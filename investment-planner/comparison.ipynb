### COMPARISON OF FUND ALLOCATION BY 1/P^2 OPTIMAL VS 1/P WEIGHTED OVER A PERIOD OF 12 STEPS

#### How to Interpret the Chart

1. Top Panel (Price) → Shows how the asset (e.g., NIFTY) moved over 12 months.
2. Bottom Panel (Bars) → Shows how much each strategy allocates per month:
3. Green (1/P) — invests more when price is low but with linear sensitivity.
4. Red (1/P²) — invests much more when price is low and very little when price is high — a quadratic emphasis.
5. So if prices dip early, the red bars will tower much higher at the left — indicating front-loaded buying.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Synthetic monthly price path (NIFTY-like pattern) ---
months = np.arange(1, 13)
prices = np.array(
    [22000, 21500, 21000, 20500, 21000, 21500, 22500, 23000, 23500, 24000, 24500, 25000]
)

# --- Total investment (W) ---
W = 12000

# --- 1/P and 1/P² weights ---
w_1P = 1 / prices
w_1P2 = 1 / (prices**2)

# --- Normalize to total capital W ---
I_1P = W * (w_1P / np.sum(w_1P))
I_1P2 = W * (w_1P2 / np.sum(w_1P2))

# --- Plotting ---
plt.figure(figsize=(10, 6))

# Upper subplot: Price trend
plt.subplot(2, 1, 1)
plt.plot(months, prices, color="black", marker="o", linewidth=2, label="Price (NIFTY)")
plt.title("Price Trend and Allocation Comparison", fontsize=13)
plt.ylabel("Price Level")
plt.grid(True, alpha=0.4)
plt.legend()

# Lower subplot: Allocation comparison
plt.subplot(2, 1, 2)
plt.bar(
    months - 0.15, I_1P, width=0.3, label="1/P Weighted", color="#2ca02c", alpha=0.8
)
plt.bar(
    months + 0.15, I_1P2, width=0.3, label="1/P² Optimal", color="#d62728", alpha=0.8
)
plt.xlabel("Month")
plt.ylabel("Investment Allocation (₹)")
plt.title("Monthly Investment Allocations under Two Strategies", fontsize=12)
plt.grid(True, axis="y", alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()
```
