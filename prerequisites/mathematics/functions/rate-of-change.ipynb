### AVERAGE RATE OF CHANGE

You have decided to buy a new Honda Accord LX, but you are concerned about the value of
the car depreciating over time. You search the Internet and obtain the following information.

* Suggested retail price $20,905
* Depreciation per year $1750 (assume constant)

Reference: Mathematics in Action: Algebric, Graphical, and Trigonometric Problem Solving. <br/>
Activity: 1.7 Depreciation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
initital_value_of_car = 20905

years = np.arange(0, 9)
car_values = [20905] * len(years)

for yi, year in enumerate(years):
    car_values[yi] -= 1750 * yi

plt.plot(years, car_values)
plt.axis([0, 8, 0, 25000])
plt.xlabel("years of ownership")
plt.ylabel("value of car")
plt.title("Car value depreciation")
plt.axis([-2, 16, -500, 25000])

index = np.argmax(np.abs(car_values))
intercept_point = (years[index], car_values[index])
print(f"The line intercepts y-axis at {car_values[index]} where x is {years[index]}.")
print(f"Vertical intercept coordinates: ({years[index]}, {car_values[index]})")

plt.show()
```

Identify the slope and vertical intercept of the line whose equation is given. <br/>
Write the vertical intercept as an ordered pair:
* $y=-2x+5$
* $s = \frac{3}{4}t+2$
* $q=2-r$
* $y=\frac{5}{6}+\frac{x}{3}$

```python
x = np.arange(-5, 5)
y = -2 * x + 5
plt.plot(x, y, label="$y=-2x+5$")

t = x
s = 3 / 4 * t + 2  # PEMDAS
plt.plot(t, s, label="$s=\\frac{3}{4}t + 2$")

r = x
q = 2 - r
plt.plot(r, q, label="$q=2-r$")

y = 5 / 6 + x / 3
plt.plot(x, y, label="$y=\\frac{5}{6}+\\frac{x}{3}$")

plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
