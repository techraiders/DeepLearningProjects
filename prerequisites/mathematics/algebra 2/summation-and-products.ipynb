### SUMMATION AND PRODUCTS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst = [1, 3, 4, 1, 6]
print(sum(lst))
print(np.sum(lst))
```

```python
print(np.prod(lst))
print(np.cumsum(lst))
print(np.cumulative_sum(lst))

plt.plot(lst, "rs-")
plt.plot(np.cumsum(lst), "bo-")

plt.legend(["set", "cumsum"])
plt.show()
```

### EXERCISE

```python
# 1) SUMMATION

a = np.arange(1, 6)  # [1 2 3 4 5]

prod = np.sum(a)
prod2 = np.sum(a**2)

ans1 = prod / prod2
ans2 = 1 / prod

print("Option 1 = %s \nOption 2 = %s" % (ans1, ans2))
```

```python
# 2) PRODUCTS

prod = np.prod(a)
prod2 = np.prod(a**2)

ans1 = prod / prod2
ans2 = 1 / prod

print("Option 1 = %s \nOption 2 = %s" % (ans1, ans2))
```
