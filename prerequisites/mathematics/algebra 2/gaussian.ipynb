#### FIND A SPECIFIC POINT ON A GAUSSIAN

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# gaussian
t = np.linspace(-3, 3, 1001)
h = 0.999

g = np.exp(-4 * np.log(2) * t**2 / h**2)

plt.plot(t, g)
plt.xlabel("t")
plt.ylabel("g")
plt.title("Gaussian with FWHM= " + str(h))

plt.show()
```

```python
yval2find = 0.5
tvals = h * np.sqrt(np.log(yval2find) / (-4 * np.log(2)))
print(tvals)
```

#### EXERCISE

```python
N = 100
G = np.zeros((N, len(t)))
h = np.zeros(N)

for i in range(N):
    h[i] = (i + 1) / N
    G[i, :] = np.exp(-4 * np.log(2) * t**2 / h[i] ** 2)

plt.pcolormesh(t, h, G)
plt.xlabel("t")
plt.ylabel("h")
plt.show()
```
