### LOG SPACED AND LINEARLY SPACED NUMBERS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.linspace(1, 2, 10)
```

```python
a = np.log10(1)
b = np.log10(10)
nums = np.logspace(a, b, 10)
print(nums)
plt.plot(nums)
plt.show()
```

```python
# Demonstrates:
#   - The difference between "log-spaced" and "linear-spaced" numbers.
#   - Linear functions are good local approximations of non-linear functions when difference is very small.

a = 40
b = 45
n = 50

li = np.linspace(a, b, n)
lo = np.logspace(np.log10(a), np.log10(b), n)

plt.plot(li, label="linear")
plt.plot(lo, label="log")
plt.legend()
plt.axis("square")

plt.show()
```
