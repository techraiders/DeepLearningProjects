### EVEN AND ODD FUNCTION

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# EVEN FUNCTION
x = np.linspace(-5, 5, 20)

f_even = x**2
f_even_neg = (-x) ** 2

plt.plot(x, f_even)
plt.plot(x, f_even_neg, "ro")
plt.show()
```

```python
f_odd = x**3
f_odd_neg = (-x) ** 3
f_neg_odd = -f_odd

plt.plot(x, f_odd)
plt.plot(x, f_odd_neg, "ro")
plt.plot(x, f_neg_odd, "g")

plt.show()
```

```python
# COS THETA IS AN EVEN FUNCTION

theta = np.linspace(-2 * np.pi, 2 * np.pi, 39)
cosfun = np.cos(theta)
cosfun_neg = np.cos(-theta)

plt.plot(theta, cosfun)
plt.plot(theta, cosfun_neg, "ro")
plt.title("$cosine(\\theta)$")

plt.show()
```

```python
# SIN THETA IS AN ODD FUNCTION

theta = np.linspace(-2 * np.pi, 2 * np.pi, 39)
sinfun = np.sin(theta)
sinfun_neg = np.sin(-theta)

plt.plot(theta, sinfun)
plt.plot(theta, sinfun_neg, "ro")
plt.title("$sin(\\theta)$")

plt.show()
```
