### MAXIMA AND MINIMA OF A FUNCTION

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst = [-4, 5.5, 1, 2]
maxval = np.max(lst)
maxvalidx = np.argmax(lst)
lst[maxvalidx]
```

```python
x = np.linspace(0, 2 * np.pi, 20)
fx = -(np.cos(x) + np.sqrt(x))

fmax = np.max(fx)
fmaxidx = np.argmax(fx)
print(fmax, fmaxidx)

plt.plot(x, fx, "bo-")
plt.plot(x[fmaxidx], fmax, "rs")
plt.show()
```

```python
x = np.linspace(-2, 2, 51)
fx = x**3 + x**4
fminidx = np.argmin(fx)

plt.plot(x, fx, "bs-")
plt.plot(x[fminidx], fx[fminidx], "rs")

plt.xlim([-1.2, 0.5])
plt.ylim([-0.2, -0.03])

plt.show()
```
