```python
v = np.arange(0, 11)
print(v)
print(np.diff(v))
print(np.diff(v, 2))
print(np.diff(np.diff(v)))
```

```python
x = np.linspace(-2, 2, 141)
fx = x**2

dx = np.diff(x)
df = np.diff(fx)

dfbydx = df / dx

plt.plot(x, fx, label="f(x)")
plt.plot(x[:-1], dfbydx, label="df(x)/dx")
plt.axis([-2, 2, -1, 2])
plt.legend()
plt.grid()

plt.show()
```
