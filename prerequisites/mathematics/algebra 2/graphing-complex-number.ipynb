### GRAPHING COMPLEX NUMBERS

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
z = complex(2, 3)
plt.plot(np.real(z), np.imag(z), "ro")
plt.plot([0, np.real(z)], [0, np.imag(z)], "r")

plt.title("COMPLEX PLANE")
plt.xlabel("real")
plt.ylabel("imag")
plt.grid()
plt.axis([-4, 4, -4, 4])
plt.show()
```

#### EXERCISE:

```python
z1 = complex(-3, 1)
z2 = complex(-1, 1)

z1pz2 = z1 + z2

plt.plot([0, np.real(z1)], [0, np.imag(z1)], label="z1")
plt.plot([0, np.real(z2)], [0, np.imag(z2)], label="z2")
plt.plot([0, np.real(z1pz2)], [0, np.imag(z1pz2)], label="z1+z2")
plt.grid()
plt.legend()
plt.axis("square")
plt.axis([-5, 5, -5, 5])
plt.show()
```

```python
z1 = complex(-3, 1)
z2 = complex(-1, 1)

z1tz2 = z1 * z2

plt.plot([0, np.real(z1)], [0, np.imag(z1)], label="z1")
plt.plot([0, np.real(z2)], [0, np.imag(z2)], label="z2")
plt.plot([0, np.real(z1tz2)], [0, np.imag(z1tz2)], label="z1*z2")
plt.grid()
plt.legend()
plt.axis("square")
plt.axis([-5, 5, -5, 5])
plt.show()
```
