```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# UNIFORM DISTRIBUTION
nums = np.random.rand(1000)
nums += 3  # shifting random distribution to generate numbers between 3 and 4.

plt.plot(nums, "s")
plt.show()

plt.hist(nums)
plt.show()
```

```python
# Restricting minimul and maximum value of a random distribution

nums = np.random.rand(1000)

minval = 2
maxval = 17

nums = nums * (maxval - minval)  # i.e. generates random numbers between 0 and 15

plt.plot(nums, "s")
plt.show()

plt.hist(nums)
plt.show()
```

```python
# NORMAL DISTRIBUTION

nums = np.random.randn(1000)
plt.plot(nums, "s", alpha=0.5)
plt.show()

plt.hist(nums)
plt.show()
```

```python
# Generate a normal distribution with mean = 15, standard deviation = 4.3

desired_mean = 15
desired_std = 4.3

nums = np.random.randn(1000)
nums = nums - np.mean(nums)
nums = nums / np.std(nums)
nums = nums * desired_std + desired_mean

plt.subplot(121)
plt.plot(nums, "s")
# plt.show()

plt.subplot(1, 2, 2)
plt.hist(nums)
plt.show()

print(
    "This distribution has a mean of %s and a standard deviation of %s"
    % (np.mean(nums), np.std(nums))
)
```
