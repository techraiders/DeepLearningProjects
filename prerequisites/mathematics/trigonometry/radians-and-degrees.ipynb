### RADIANS AND DEGREES

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deg = 180

print("%g degrees is %g radians." % (deg, deg * np.pi / 180))
```

```python
deg = 180  # spiral i.e. looping around circle many times
print("%g degrees is %g radians." % (deg, deg * np.pi / 180))

# Convert the degrees back into a number between 0 and 2pi to find the degree

rad = (
    deg * np.pi / 180
)  # 180 degrees equals to 3.14 radians where 3.14 equals to np.pi,
# so given degree * np.pi/180 calculates to x radian by unity method

rad = rad % (2 * np.pi)  # find the remainder to figure out how
print("%g degrees is %g radians." % (deg, rad))
```

```python
rad = np.deg2rad(deg)
print("%g degrees is %g radians." % (deg, rad))

deg = np.rad2deg(rad)
print("%g radians is %g degrees." % (rad, deg))
```

#### EXERCISE

```python
def angle_convert_plot():

    # get the inputs from the user
    angle = eval(input("What angle to convert?"))
    unit = input("Which unit did you input (radian or degree)")

    if unit[0].lower() == "r":
        rad = angle
        deg = np.rad2deg(angle)
    elif unit[0].lower() == "d":
        deg = angle
        rad = np.deg2rad(deg)
    else:
        raise ValueError("Unknown unit!")

    # convert to < 360
    deg = deg % 360
    rad = rad % (2 * np.pi)

    # make the graph
    plt.plot([0, 1], [0, 0], "r", linewidth=2)
    plt.plot([0, np.cos(rad)], [0, np.sin(rad)], "r", linewidth=3)
    plt.axis("square")
    plt.axis([-1, 1, -1, 1])
    plt.grid()
    plt.title("Angle is %s$^o$ or %s rad." % (deg, rad))
```

```python
angle_convert_plot()
```
