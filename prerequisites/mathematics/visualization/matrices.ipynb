#### MAKING IMAGES FROM MATRICES

```python
import numpy as np
import matplotlib.pyplot as plt

A = [[1, 2], [1, 4]]
plt.xticks([0, 1])
plt.yticks([0.85, 1.04])
plt.imshow(A)
plt.show()
```

```python
A = np.zeros((10, 14), dtype=int)

for i in range(0, np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        A[i, j] = 3 * i - 4 * j
plt.imshow(A)
plt.plot([0, 4], [8, 2], "r", linewidth=2)

for i in range(0, np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        plt.text(
            j, i, A[i, j], horizontalalignment="center", verticalalignment="center"
        )

# plt.set_cmap("Purples")
plt.show()
```

#### EXERCISES

```python
N = 10
C = np.zeros((N, N))

for i in range(0, N):
    for j in range(0, N):
        C[i, j] = (-1) ** (i + j)
plt.imshow(C)
plt.set_cmap("gray")
plt.tick_params(labelleft=False, labelbottom=False)
plt.show()
```
