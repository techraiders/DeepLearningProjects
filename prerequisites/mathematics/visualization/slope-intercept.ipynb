```python
import matplotlib.pyplot as plt
import numpy as np

# y = mx + b

x = [-5, 5]
m = -0.3
b = 0.1

y = [0, 0]  # initialize with any random value

# for i in range(0, len(x)):
#     y[i] = m * x[i] + b  # without numpy

y = m * np.array(x) + b  # with using numpy

plt.plot(x, y, label="y=%sx+%s" % (m, b))
plt.axis("square")
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], "k--")
plt.plot([0, 0], axis.get_ylim(), "k--")
plt.legend()

plt.show()
```
