#### DRAWING PATCHES WITH POLYGON

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

y = np.array([[1, 1], [2, 3], [3, 1]])
p = Polygon(y, facecolor="m", alpha=0.4)

y1 = np.array([[2, 2], [2.5, 4], [3.5, 1]])
p1 = Polygon(y1, facecolor="w", alpha=0.4, edgecolor="k")

fig, ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])

plt.show()
```

#### EXERCISE

```python
x = np.linspace(-2, 2, 101)
f = -(x**2)

pnts = np.vstack((x, f)).T
print(np.shape(pnts))
p = Polygon(pnts, facecolor="g", alpha=0.2)

y = np.array([[-0.5, -4], [-0.5, -2.5], [0.5, -2.5], [0.5, -4]])
p1 = Polygon(y, facecolor="w")


fig, ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

plt.plot(x, f)
plt.plot([-2, 2], [-4, -4], "k")
plt.axis("off")
plt.show()
```
