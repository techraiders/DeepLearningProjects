#### VISUALIZATIONS OF SYMBOLIC FUNCTIONS USING SYMPY

```python
import sympy as sym
import sympy.plotting.plot as symplot
from sympy.abc import x

y = x**2
p = symplot(y, show=False)
p.ylim = [0, 50]
p[0].line_color = "m"

p.show()
```

```python
x, a = sym.symbols("x,a")
expr = a / x
n = 1

p = symplot(expr.subs(a, n), (x, -5, 5), show=False)

p[0].label = "$y=%s$" % sym.latex(expr.subs(a, n))
p.legend = True
p.ylim = [-5, 5]

p.extend(symplot(expr.subs(a, 2), show=False))
p[1].label = "$y=%s$" % sym.latex(expr.subs(a, 2))

p.show()
```

#### EXERCISES

```python
y = a / (x**2 - a)

p = None
p = symplot(y.subs(a, 1), (x, -5, 5), show=False)
p[0].label = "$f(x)=y=%s$" % sym.latex(y.subs(a, 1))

for i in range(2, 5):
    p.extend(symplot(y.subs(a, i), (x, -5, 5), show=False))
    p[i - 1].label = "$f(x)=y=%s$" % sym.latex(y.subs(a, i))

p.ylim = [-10, 10]
p.legend = True
p.show()
```
