#### VISUALIZATIONS OF SYMBOLIC FUNCTIONS USING SYMPY

```python
import sympy as sym
import sympy.plotting.plot as symplot
from sympy.abc import x

y = x**2
plt1 = symplot(y, show=False)
plt1.ylim = [0, 50]
plt1[0].line_color = "m"

plt1.show()
```

```python
x, a = sym.symbols("x,a")
expr = a / x
n = 1

plt1 = symplot(expr.subs(a, n), (x, -5, 5), show=False)

plt1[0].label = "$y=%s$" % sym.latex(expr.subs(a, n))
plt1.legend = True
plt1.ylim = [-5, 5]

plt2 = symplot(expr.subs(a, 2), show=False)
plt1.extend(plt2)
plt1[1].label = "$y=%s$" % sym.latex(expr.subs(a, 2))

plt1.show()
```
