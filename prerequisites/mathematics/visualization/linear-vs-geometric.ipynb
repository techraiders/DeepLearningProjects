```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Initial setup
top_price = 100
periods = np.arange(0, 10, 1)

# Linear fall: decrease by fixed amount per period
linear_fall = top_price - 10 * periods

# Geometric fall: decrease by fixed percentage (-10%) per period
geometric_fall = top_price * (0.9) ** periods

# plt.show()

bottom_price = 50

# Linear Rise: increase by fixed amount per period
linear_rise = bottom_price + 10 * periods

# Geometric rise: increase by fixed percentage (+10%) per period
geometric_rise = bottom_price * (1.1) ** periods

# Create subplots - returns figure and axes objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Left plot
ax1.plot(periods, linear_fall, label="Linear Fall (10/period)", marker="o")
ax1.plot(periods, geometric_fall, label="Geometric Fall (10%/period)", marker="s")
ax1.set_title("Comparison: Linear vs Geometric Fall of Asset Price")
ax1.set_xlabel("Time (Periods)")
ax1.set_ylabel("Price ()")
ax1.legend()
ax1.grid(True)

# Right plot
ax2.plot(periods, linear_rise, label="Linear Rise (10/period)", marker="o")
ax2.plot(periods, geometric_rise, label="Geometric Rise (10%/period)", marker="s")
ax2.set_title("Comparison: Linear vs Geometric Rise in Asset Price")
ax2.set_xlabel("Time (Periods)")
ax2.set_ylabel("Price ()")
ax2.legend()
ax2.grid(True)

plt.show()
```
