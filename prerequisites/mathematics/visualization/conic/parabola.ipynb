```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# HORIZONTAL PARABOLA
# parameters
a = 1  # 1/4p
h = 1
k = -2
n = 100

# x-axis points to evaluate the function
x = np.linspace(h - 5, h + 5, n)

y = a * (x - h) ** 2 + k

plt.plot(x, y)
plt.grid()
plt.axis("square")
plt.show()
```

```python
# VERTICAL PARABOLA
# parameters
a = 1  # 1/4p
h = 1
k = -2
n = 100

# x-axis points to evaluate the function
x = np.linspace(h - 5, h + 5, n)

y = a * (x - h) ** 2 + k

plt.plot(y, x)
plt.grid()
plt.axis("square")
plt.show()
```

```python
# HORIZONTAL PARABOLA
# parameters
a = 1  # 1/4p
p = 1 / (4 * a)
h = 1
k = -2
n = 100

# x-axis points to evaluate the function
x = np.linspace(h - 2, h + 2, n)

# create the function
y = a * (x - h) ** 2 + k

# plot the parabola
plt.plot(x, y, label="Parabola")

# plot the vertex
plt.plot(h, k, "ro", label="Vertex")

# plot the focus
plt.plot(h, k + p, "go", label="Focus")

# plot the directix
d = k - p
plt.plot(x[[0, -1]], [d, d], label="Diretix")

plt.grid()
plt.legend()
plt.show()
```
