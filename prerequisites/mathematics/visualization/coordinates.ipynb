```python
import matplotlib.pyplot as plt

x = 3
y = 5

plt.axis("square")  # square shape plot
plt.axis([-6, 6, -6, 6])  # -6,6 i.e. x-axis range; -6, 6 i.e. y-axis range
plt.grid()
plt.plot(x, y, "go")
plt.show()
```

```python
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2, -3, 4]

for i in range(0, len(x)):
    plt.plot(x[i], y[i], "o", label="point %s" % (i + 1))

plt.axis("square")
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(4, 2, "r+")
axis = plt.gca()
ylim = axis.get_ylim()
print(ylim)
axis.set_ylim([ylim[0], 5.3984588])

plt.show()
```

```python
# 1) y = x**2 - 3x
import sympy as sym
from IPython.display import display, Math

x = sym.symbols("x")
y = x**2 - 3 * x

for pnti in range(-11, 12):
    plt.plot(pnti, y.subs(x, pnti), "o")
plt.xlabel("x")
plt.ylabel("f(x) = %s" % sym.latex(sym.sympify(y)))
plt.show()
```
