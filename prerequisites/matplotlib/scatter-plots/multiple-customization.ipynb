#### Combining Multiple Customizations

```python
# Now, let's put everything together into a single plot

import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)  # Generate random x-coordinates
y = np.random.rand(50)  # Generate random y-coordinates

plt.figure(figsize=(8, 5))  # Define figure size

# Generate random properties for points
sizes = np.random.randint(100, 300, size=50)  # Varying sizes
colors = np.random.rand(50)  # Different colors

# Create scatter plot with customizations
plt.scatter(
    x,
    y,
    s=sizes,
    c=colors,
    cmap='coolwarm',
    alpha=0.6,
    edgecolors='green',
    linewidths=1.2
)

plt.xlabel('X-axis Label', fontsize=12)
plt.ylabel('Y-axis Label', fontsize=12)
plt.title('Customized Scatter Plot', fontsize=16)
plt.colorbar(label='Color Scale')

plt.show()
```
