```python
import matplotlib.pyplot as plt
import numpy as np

x_large = np.random.rand(5000)
y_large = np.random.rand(5000)

plt.figure(figsize=(10, 6))

# Reduce marker size and increase transparency to handle large data
plt.scatter(x_large, y_large, s=5, alpha=0.2, color="blue")
plt.title("Scatter Plot with Large Dataset")

plt.show()
```
