```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data points
x = np.random.rand(
    20
)  # Generates an array of 20 random numbers between 0 and 1 for x-coordinates
y = np.random.rand(
    20
)  # Generates an array of 20 random numbers between 0 and 1 for y-coordinates

# Creating the scatter plot
plt.figure(
    figsize=(8, 5)
)  # Generates a new figure with a specific size (8 inches by 5 inches)

# Scatter plot of the random points:
#   - `s=100` sets the market size
#   - `c='red'` colors the markers red
#   - `alpha=0.6` makes the markers slightly transparent (60% opacity)
plt.scatter(x, y, s=100, c="red", alpha=0.6)

# Adding annotations to selected points
for i in range(0, len(x), 10):  # Iterate over the points, labeling every 10th point
    plt.annotate(
        f"({x[i]: .2f}, {y[i]: .2f})",  # Format the annotation to display coordinates with two decimal places
        (
            x[i],
            y[i],
        ),  # The coordinates of the point to annodate (same as the point being labeled)
        textcoords="offset points",  # Specifies the coordinates system for the text position
        xytext=(
            8,
            -11,
        ),  # Offset the text 10 pixel to the right and 10 pixels down for better visibility
        ha="center",  # Horizontally align the text at the center
    )

plt.show()
```
