```python
import matplotlib.pyplot as plt
import numpy as np

# Generate different sizes for markers
sizes = np.random.randint(100, 500, size=50) # Random sizes between 50 and 500
x = np.random.rand(50)
y = np.random.rand(50)

# Create scatter plot with varying marker sizes
plt.scatter(x, y, s=sizes, color='blue', alpha=0.5)

# Display plot
plt.show()
```
