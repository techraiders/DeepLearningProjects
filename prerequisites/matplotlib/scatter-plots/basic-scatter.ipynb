##### Creating a basic scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
np.random.seed(50) # ensures reproducibility of random numbers generated by NumPy.

# 50 is an arbitary number chosen as the seed.
# It ensures that random functions (like np.random.rand()) produce the same results every time the code runs

# Generate random data for X and Y coordinates
x = np.random.rand(50) # 50 random numbers between 0 and 1
y = np.random.rand(50)

# Create a basic scatter plot
plt.scatter(x, y)

# Display the plot
plt.show()
```
