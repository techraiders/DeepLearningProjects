```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 5))  # Define figure size
x = np.random.rand(50)  # Generate random x-coordinates
y = np.random.rand(50)  # Generate random y-coordinates
plt.scatter(
    x,
    y,
    marker="o",  # Set the marker type e.g. 'o', 's', 'D', '^', 'v', '*', 'x'
    s=100,  # Set the marker size
    color="b",  # Set the marker color e.g. 'b', 'g', 'r', 'c', 'm', 'y', 'k
)
plt.show()
```
