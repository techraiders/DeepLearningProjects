#### Multiple lines with Different Customizations
Plot the following functions of the same graph
* y = x^3 (solid, red, square markers)
* y = (-2x^2) + 5 (dashed, purple, triangle, markers)
* y = 3x-4 (dotted, orange, cross markers)

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from -3 to 30 points
x = np.linspace(-3, 3, 30)

# Define the function

y1 = x**3
y2 = -2*x**2 + 5
y3 = 3*x - 4

# Plot the first function (solid line, red, square markers)
plt.plot(x, y1, linestyle="-", color='red', marker='s', label='y = x^3')
plt.plot(x, y2, linestyle="--", color='purple', marker='v', label='y = -2x^2 + 5')
plt.plot(x, y3, linestyle=":", color='orange', marker='x', label='y = 3x - 4')

# Adding labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
